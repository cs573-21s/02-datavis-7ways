```
import scipy
from scipy.io import loadmat
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


carData = pd.read_csv('/content/drive/My Drive/cars-sample-updated.csv') 
df=pd.DataFrame(carData)
```

```
x=df['Weight']
y=df['MPG']
label = df['Manufacturer']
groups = df.groupby(label)
# Plot
fig, ax = plt.subplots()
ax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
for name, group in groups:
  weight =  group.Weight.values
  mpg = group.MPG.values
  plt.scatter(weight, group.MPG, marker='o', s=weight/40, label=name, alpha=0.5)#linestyle='', label=name)
plt.legend()
plt.title("Python")
plt.xlabel("Weight")
plt.ylabel("MPG")
plt.show()
```
