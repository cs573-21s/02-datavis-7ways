```python
import warnings
warnings.filterwarnings("ignore")

%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import lines, rcParams, cycler
import numpy as np, random, operator, pandas as pd
import copy
from cycler import cycler
import scipy.stats as stats
```

```python
carsample = pd.read_csv('cars-sample.csv')
```

```python
x = carsample['Weight']
y = carsample['MPG']

plt.rc('axes', prop_cycle=(cycler('color', ['red', '#7c8531', '#22c785', 'c', 'magenta'])))

groups = carsample.groupby("Manufacturer")

for name, group in groups:
    plt.scatter(group["Weight"], group["MPG"], alpha=0.5, s=(group["Weight"])**2*0.000005)
plt.grid(True)
plt.xlabel("Weight")
plt.ylabel("MPG")
plt.show()
```
