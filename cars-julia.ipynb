# Julia and Vega Lite

```julia
using Pkg
using VegaLite, VegaDatasets
using CSV
using DataFrames
# using Vega
```

```julia
data = CSV.read("data/cars_cleaned.csv", DataFrame)
```

```julia
data |> @vlplot(
    config={view={stroke=nothing}},
    title={
        text="Weight vs MPG", 
        fontSize=20, 
        subtitle="Julia and Vega-Lite", 
        subtitleFontSize=15,
        subtitlePadding=10
    },
    mark={
        :circle,
        opacity=0.5
    },
    selection={
        view={type=:interval, bind=:scales} # zoom in and out
    },
    x={:Weight, axis={grid=false}, scale={domain=(1250, 5250)}}, 
    y={:MPG, axis={grid=false}, scale={domain=(0, 50)}},
    color=:Manufacturer, 
    size=:WeightCategory,
    width=500,
    height=500
)
```
