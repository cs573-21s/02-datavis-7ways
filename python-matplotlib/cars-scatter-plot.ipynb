```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
#import data
cars = pd.read_csv("cars-sample.csv")
```

```python
#cars
```

```python
color = {'bmw':'#E87D72', 'ford':'#A3A433', 'honda':'#56BB82', 'mercedes':'#4DAEF0', 'toyota':'#D875EC'}
x_ticks = [2000, 3000, 4000, 5000]
y_ticks = [10, 20, 30, 40]
```

```python
plt.style.use('ggplot')
fig, ax = plt.subplots()
ax.scatter(cars.Weight, cars.MPG, s= cars['Weight']*0.035, c = cars['Manufacturer'].map(color), alpha=0.6);
ax.set(xlabel = "Weight",
      ylabel = "MPG",
      title= "MPG and Wights of Cars");
ax.set_xticks(x_ticks);
ax.set_yticks(y_ticks);
```
